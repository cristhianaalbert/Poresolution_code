```python
def count_alpha_error (a, alpha=0.1):

    '''Calculates the percentage of elements inside a +- alpha range around the true measured value of the ion;
     a = array with final predictions on unseen data;
     alpha = range of error
    '''
    error = a.OHpredicted / a.OH                           #when we are dealing with log scale of ion, the analysis is still in absolute terms;
    count = error[((1-alpha) < error) & (error <(1+alpha))].size
    total = len(error)
    return count/total
```

# Feature importance scores 


Feature importance scores in simple format

```python
#creating a list with the names of the models and features, without repeating them. It will be used to define the columns on the dataframe;
count_model_column = []
count_featimp_column = []
for i in range(totalsystems):
  modelnamec = finaltable.iloc[i,0]  
  if count_model_column.count(modelnamec)==0:    #feature has not been counted until now;
    count_model_column.append(modelnamec)
  for j in range(10,10+2*mostimpfeat,2):

    feat = finaltable.iloc[i,j]
    value = finaltable.iloc[i,j+1]
    
    if count_featimp_column.count(feat)==0:    #feature has not been counted until now;
      count_featimp_column.append(feat)
 
#creating the dataframe to store the values of the most important features, consulting all the "finaltable"
count_featimp_score = pd.DataFrame(np.zeros(shape=(totalsystems,len(count_featimp_column))))
count_featimp_score.columns = [count_featimp_column]
for i in range(totalsystems):
  count_featimp_score.iloc[i,0] = finaltable.iloc[i,0]
  for j in range(10,10+2*mostimpfeat,2):

    feat = finaltable.iloc[i,j]
    value = finaltable.iloc[i,j+1]
    index = count_featimp_column.index(feat)
    count_featimp_score.iloc[i,index] = value

#SOLVING THE PROBLEM WITH DELETING FIRST COLUMN OF MOST IMPORTANT FEATURE. 
#This happened when the most imp feat for the first model didn't appear in one of the other models. 
#Instead of giving it the value zero, the name of the model was appearing. 
#So, when summing all values for the next step (when putting them in order), this feature could not be summed and it would disappear from the list.
#For the other systems, there is not problem with not having a feature. It only happened for the first one (I don't know why).
#Its a simple trick but it works. 
  for modelnamed in model_list:
    if count_featimp_score.iloc[i,0]==modelnamed:
      count_featimp_score.iloc[i,0]=0

count_featimp_score.index=finaltable.iloc[:,0]

#organizing the finalorder of features and sorting them.
#finalorder = count_featimp_score.sum().sort_values(ascending=False)

error = count_featimp_score.std().div(count_featimp_score.mean().max())
error = np.array(error).tolist()
finalorderassist = pd.DataFrame(count_featimp_score.mean())
finalorderassist.at[:,1]=error                    
finalorderassist.columns = ['Mean','Std']
finalorderassist.sort_values(by='Mean',ascending=False, inplace=True)

finalorder=pd.DataFrame(finalorderassist['Mean'])                 
error = pd.DataFrame(finalorderassist['Std'])


#finalorder = pd.DataFrame(finalorder)
finalorder_norm = finalorder.div(finalorder.iloc[0,0])
error_norm = error
finalorder_norm.columns = ['Feature Importance']
finalorder_norm.to_csv('finalorder_norm.csv',index=True)
display(finalorder_norm)

count_featimp_score.to_csv('countfeatimpscoretest.csv',index=True)

plt.rcdefaults()
fig, ax = plt.subplots(figsize=(8,4.5),dpi=600)
finalorder_norm.iloc[0:20].plot.bar(yerr=error_norm.iloc[:20,0],ax=ax,color='Gray')
ax.set_ylabel('Feature importance')
plt.show()
```
